## 1. Scatterplot
- 산점도를 표시해 주는 그래프

0 ~ 1 사이의 임의의 랜덤한 값을 생성합니다. 

```python
np.random.rand(50)
```

0 부터 50개의 값을 순차적으로 생성합니다.

```python
np.arange(50)
```

### 1.1. x, y, colors, area 설정하기

* colors 는 임의 값을 color 값으로 변환합니다.
* area는 점의 넓이를 나타냅니다. 값이 커지면 당연히 넓이도 커집니다.

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)
area = x*y*1000
area 
```

```python
#plt.scatter(x,y)
plt.scatter(x,y, s=area, c=colors)
plt.scatter(x,y, s=area, c=colors, cmap='plasma')

plt.colorbar()
plt.show()
```

### 1.2. cmap과 alpha

* cmap에 컬러를 지정하면, 컬러 값을 모두 같게 가져갈 수도 있습니다.
* alpha값은 투명도를 나타내며 0 ~ 1 사이의 값을 지정해 줄 수 있으며, 0에 가까울 수록 투명한 값을 가집니다.

```python
plt.figure(figsize=(12,6))

plt.subplot(131)
plt.scatter(x,y, s=area, cmap='Blues', alpha=0.1)
plt.title('alpha= 0.1')

plt.subplot(132)
plt.scatter(x,y, s=area, cmap='Blues', alpha=0.5)
plt.title('alpha= 0.5')

plt.subplot(133)
plt.scatter(x,y, s=area, cmap='Blues', alpha=1.0)
plt.title('alpha= 1.0')

plt.show()
```

### Batplot에서 비교 그래프 그리기

```python
# 두학생의 그래프
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))
# print(x)
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

# 넓이 지정
bar_width = 0.35  # 비율 (절대값X)

# xtick 설정
x1 = np.arange(len(y_1))
x2 = [i + bar_width for i in x1]  # 막대넓이만큼 더해줘서 바로 옆에 붙어있는 형태 만들어줌.
plt.xticks(x1+(bar_width/2), x_label)

# 그래프 그리기
plt.bar(x1, y_1, width=bar_width)
plt.bar(x2, y_2, width=bar_width)

plt.ylabel('Score of Subject')
plt.title('Subject')
plt.legend(['bob', 'andy'])   #범례

plt.show()
```

## 3. Line Plot

### 3.1. 기본 lineplot 그리기

```python
x = np.arange(0,10,0.1)
y = 1 + np.sin(x)

plt.plot(x,y)

plt.xlabel('x value',fontsize = 12)
plt.ylabel('y value',fontsize = 12)
plt.title('sin graph', fontsize=15)

plt.grid()
plt.show()
```

### 3.2. 2개 이상의 그래프 그리기

* color: 컬러 옵션
* alpha: 투명도 옵션

```python
x = np.arange(0,10,0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x,y_1, label = '1+sin', color='blue', alpha=0.3)
plt.plot(x,y_2, label = '1+cos', color='red', alpha=0.7)

plt.xlabel('x value',fontsize = 12)
plt.ylabel('y value',fontsize = 12)
plt.title('sin and cos graph', fontsize=15)

#plt.grid()
plt.legend()

plt.show()
```

### 3.3. 마커 스타일링

* marker: 마커 옵션

```python
x = np.arange(0,10,0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x,y_1, label = '1+sin', color='blue', alpha=0.3, marker = 'o', markersize = 3)
plt.plot(x,y_2, label = '1+cos', color='red', alpha=0.7, marker = '*', markersize = 3)

plt.xlabel('x value',fontsize = 12)
plt.ylabel('y value',fontsize = 12)
plt.title('sin and cos graph', fontsize=15)

#plt.grid()
plt.legend()

plt.show()
```

### 3.4. 라인 스타일 변경하기

* linestyle: 라인 스타일 변경 옵션

```python
x = np.arange(0,10,0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x,y_1, label = '1+sin', color='blue', linestyle=':')
plt.plot(x,y_2, label = '1+cos', color='red', linestyle='-.')

plt.xlabel('x value',fontsize = 12)
plt.ylabel('y value',fontsize = 12)
plt.title('sin and cos graph', fontsize=15)

#plt.grid()
plt.legend()

plt.show()
```

## 4. Areaplot (Filled Area)

- Line Plot에서 아랫부분이 색칠된것.
- matplotlib에서 area plot을 그리고자 할 때는 **fill_between 함수**를 사용


### 4.1. 기본 areaplot 그리기

```python
x = np.arange(1,21)
y = np.random.randint(5,10,20)

plt.fill_between(x, y, color='green', alpha = 0.2)

plt.show()
```

### 4.2. 경계선을 굵게 그리고 area는 옅게 그리는 효과 적용

```python
plt.fill_between(x, y, color='pink', alpha = 0.2)

# 경계선에 한번 더 그려준다.
plt.plot(x,y, color='pink')

plt.show()
```

### 4.3. 여러 그래프를 겹쳐서 표현

```python
x = np.arange(1,10,0.05)
y_1 = np.cos(x)+1
y_2 = np.sin(x)+1
y_3 = y_1*y_2

plt.fill_between(x,y_1,color='red',alpha=0.2)
plt.fill_between(x,y_2,color='green',alpha=0.2)
plt.fill_between(x,y_3,color='blue',alpha=0.2)

plt.show()
```

## 5. Pie Chart 

- 일반적으로 점유율을 표시할 때 많이 사용(4~5개 비교시 효과 좋다.)
- patches, texts, autotexts 인자를 리턴 받는다.
- **texts**는 label에 대한 텍스트 효과
- **autotexts**는 파이 위에 그려지는 텍스트 효과를 다룰 때 활용

**pie chart 옵션**
- explode: 파이에서 툭 튀어져 나온 비율
- autopct: 퍼센트 자동으로 표기
- shadow: 그림자 표시
- startangle: 파이를 그리기 시작할 각도

```python
brand = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'Oppo', 'Etc']
rate = [20.4, 15.8, 10.5, 9, 7.6, 36.7]

# texts, autotexts 인자를 활용하여 텍스트 스타일링을 적용합니다
plt.pie(rate, labels=brand)
plt.title('Smartphone pie', fontsize=15)

plt.show()
```

```python
brand = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'Oppo', 'Etc']
rate = [20.4, 15.8, 10.5, 9, 7.6, 36.7]
rate = [50, 15, 10, 8, 5, 25]

ex = (0.3, 0, 0, 0, 0, 0)
patchs, texts, autotexts = plt.pie(rate, 
                                   explode = ex,# 조각 떨어트리는 정도
                                   labels = brand,
                                   autopct= '%0.2f%%', # 자동으로 비율계산 (소숫점 몇 자리까지 표시할건지)
                                   shadow = True,# 그림자
                                   startangle = 90)#시작하는 값의 각도

plt.title('Smartphone pie', fontsize = 15)

#print(texts)
# label 텍스트에 대한 스타일 적용
for t in texts :
    t.set_fontsize(12)
    t.set_color('gray')

# pie 위의 텍스트에 대한 스타일 적용
for t in autotexts :
    t.set_fontsize(15)
    t.set_color('white')
    
plt.show()
```

## 6. Histogram
- 데이터의 밀도를 볼때 많이 쓰이는 그래프

### 6.1. 기본 Histogram 그리기

```python
#x = np.random.randn(30) # 10만개의 랜덤한 데이터를 생성
#bins = 10 # 30구간으로 나누어 본다.
#print(x)

x = np.random.randn(100000) # 10만개의 랜덤한 데이터를 생성
bins = 30 # 30구간으로 나누어 본다.

plt.hist(x,bins=bins)
plt.show()
```

### 6.2. 다중 Histogram 그리기

* sharey: y축을 다중 그래프가 share
* tight_layout: graph의 패딩을 자동으로 조절해주어 fit한 graph를 생성

```python
x = np.random.randn(100000)
bins = 30 

fig, axes = plt.subplots(1,3, sharey=True, tight_layout=True)
fig.set_size_inches(12,5)   # 사이즈 조절

axes[0].hist(x, bins = bins)
axes[1].hist(x, bins = bins*2)
axes[2].hist(x, bins = bins*4)


plt.show()
```

### 6.3. Y축에 Density 표기
- y축에 빈도수가 아닌 %를 표기하고자 할때

```python
x = np.random.randn(100000)
bins = 30 

fig, axes = plt.subplots(1,2, tight_layout=True)
fig.set_size_inches(12,5)   # 사이즈 조절

axes[0].hist(x, bins = bins, density=True, cumulative = True) # cumulative=True : 누적분포
axes[1].hist(x, bins = bins, density=True)



plt.show()
```

## 7. Box Plot

![Box Plot](https://justinsighting.com/wp-content/uploads/2016/12/boxplot-description.png)
- image source : https://justinsighting.com/how-to-interpret-box-plots/


- IQR(Inter Quartile Range) = (3Q-1Q)*1.5
- Upper Whisker 구간 = 3Q+IQR
- Lower Whisker 구간 = 1Q-IQR
- Outliers 가 얼마나 많이 분포되어있는 알아볼때 많이 사용

```python
# 샘플 데이터 생성 
spread = np.random.rand(50) * 100 # 0~1 사이의 숫자에 100 곱한거 50개
center = np.ones(25) * 50 # 50 25개

flier_high = np.random.rand(10) * 100 + 100 # 0~1사이의 10개
flier_low = np.random.rand(10) * -100 # 10개

#nd array를 합칠수 있는 함수
data = np.concatenate((spread, center, flier_high, flier_low))
print(data)
```

### 7.1 기본 박스플롯 생성

```python
plt.boxplot(data)
plt.tight_layout()
plt.show()
```

### 7.2. Box Plot 축 바꾸기

**vert=False**  옵션을 통해 표시하고자 하는 축을 바꿀 수 있습니다.

```python
plt.title('Horizontal Box Plot', fontsize=15)
plt.boxplot(data, vert=False)
plt.show()
```

### 7.3. Outlier 마커 심볼과 컬러 변경

```python
outlier_marker = dict(markerfacecolor = 'r', marker = 'D')

plt.boxplot(data, flierprops=outlier_marker)
plt.show()
```

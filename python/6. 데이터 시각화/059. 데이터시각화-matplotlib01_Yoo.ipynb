### 1.2. 다중 그래프 (multiple graphs)

1개의 canvas 안에 다중 그래프 그리기

```python
data = np.arange(1,51)
plt.plot(data)

data2 = np.arange(51,101)
plt.plot(data2)

plt.plot(data2+50)

plt.show()
```

```python
data = np.arange(100,201)
plt.plot(data)

data2 = np.arange(200, 301)
plt.figure()
plt.plot(data2)

plt.show()
```

### 1.3. 여러개의 plot을 그리는 방법 (subplot)

subplot(row, column, index)

```python
data = np.arange(100,201)
plt.subplot(2,1,1)    # 2행 1열의 1번째 그래프라는 의미  (콤마 지워도 똑같이 나옴)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(2,1,2)
plt.plot(data2)

plt.show()
```

```python
data = np.arange(100,201)
plt.subplot(1,3,1)
plt.plot(data)

data2 = np.arange(200,301)
plt.subplot(1,3,2)
plt.plot(data2)

data3 = np.arange(300,401)
plt.subplot(1,3,3)
plt.plot(data3)

# 조금 더 타이트하고 이쁘게 출력되는 옵션
plt.tight_layout()
plt.show()
```

```python
data = np.arange(100,201)
data2 = np.arange(200,301)
data3 = np.arange(300,401)

plt.subplot(2,3,1)
plt.plot(data)

plt.subplot(2,3,2)
plt.plot(data2)

plt.subplot(2,3,3)
plt.plot(data3)

plt.subplot(2,3,4)
plt.plot(data)

plt.subplot(2,3,5)
plt.plot(data2)

plt.subplot(2,3,6)
plt.plot(data3)

# 조금 더 타이트하고 이쁘게 출력되는 옵션
plt.tight_layout()
plt.show()
```

위의 코드와 동일하나 , (콤마)를 제거한 상태

```python
data = np.arange(100,201)
plt.subplot(211)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(212)
plt.plot(data2)

plt.show()
```

### 1.4. 여러개의 plot을 그리는 방법 (subplots) - s가 더 붙습니다.

plt.subplots(행의 갯수, 열의 갯수)

```python
data = np.arange(1,51)

# 밑그림
fig, axes = plt.subplots(2,3)

axes[0,0].plot(data)
axes[0,1].plot(data*data)
axes[0,2].plot(data**3)
axes[1,0].plot(data%10)
axes[1,1].plot(-data)
axes[1,2].plot(data//20)

plt.tight_layout()
plt.show()
```

### 2.6. 스타일 세부 설정 - 마커, 라인, 컬러

스타일 세부 설정은 마커, 선의 종류 설정, 그리고 컬러가 있으며, 문자열로 세부설정을 하게 됩니다.

[세부 도큐먼트 확인하기](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

**marker의 종류**
* '.'	point marker
* ','	pixel marker
* 'o'	circle marker
* 'v'	triangle_down marker
* '^'	triangle_up marker
* '<'	triangle_left marker
* '>'	triangle_right marker
* '1'	tri_down marker
* '2'	tri_up marker
* '3'	tri_left marker
* '4'	tri_right marker
* 's '	square marker
* 'p'	pentagon marker
* '*'	star marker
* 'h'	hexagon1 marker
* 'H'	hexagon2 marker
* '+'	plus marker
* 'x'	x marker
* 'D'	diamond marker
* 'd'	thin_diamond marker
* '|'	vline marker
* '_'	hline marker

```python
plt.plot(np.arange(1,11), np.arange(1,11)*2, marker = 'o', markersize = 5)
plt.plot(np.arange(1,11), np.arange(1,11)*2-10, marker = 'v', markersize = 10)
plt.plot(np.arange(1,11), np.arange(1,11)*2-20, marker = 'D', markersize = 15)
plt.plot(np.arange(1,11), np.arange(1,11)*2-30, marker = '*', markersize = 20)

# 타이틀 & 폰트 사이즈
plt.title('마커 설정 예제', fontsize=20)


# x축, y축 label 설정
plt.xlabel('X축', fontsize=20)
plt.ylabel('Y축', fontsize=20)

#x tick, y tick  각도설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

#legend 설정
plt.legend(['X2','X2-10','x2-20', 'x33'], fontsize=15)


plt.show()

```

**line의 종류**
* '-' solid line style
* '--' dashed line style
* '-.' dash-dot line style
* ':' dotted line style

```python
plt.plot(np.arange(1,11), np.arange(1,11)*2, marker = 'o', linestyle='')
plt.plot(np.arange(1,11), np.arange(1,11)*2-10, marker = 'v', linestyle='-')
plt.plot(np.arange(1,11), np.arange(1,11)*2-20, marker = 'D', linestyle='--')
plt.plot(np.arange(1,11), np.arange(1,11)*2-30, marker = '*', linestyle='-.')
plt.plot(np.arange(1,11), np.arange(1,11)*2-40, marker = '*', linestyle=':')

# 타이틀 & 폰트 사이즈
plt.title('마커 설정 예제', fontsize=20)


# x축, y축 label 설정
plt.xlabel('X축', fontsize=20)
plt.ylabel('Y축', fontsize=20)

#x tick, y tick  각도설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

#legend 설정
plt.legend(['X2','X2-10','x2-20', 'x33'], fontsize=15)


plt.show()

```

## 3. 이미지를 저장하고 활용하기
- 다양한 그래프를 그려서 다양하게 활용(보고서, 리포트, 논문...) 하기 위해 img 저장

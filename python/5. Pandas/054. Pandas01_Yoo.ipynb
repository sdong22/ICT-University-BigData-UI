## 1.3. DataFrame
- Series가 1차원이라면 DataFrame은 2차원으로 확대된 버젼
- Excel spreadsheet이라고 생각하면 이해하기 쉬움
- 2차원이기 때문에 인덱스가 row, column로 구성됨
- row는 각 개별 데이터를, column은 개별 속성을 의미
- Data Analysis, Machine Learning에서 data 변형을 위해 가장 많이 사용

# 5. 결측값(Null) 알아보기

## 5-1. NaN 값에 대하여

* Null 값은 **비어있는 값, 결측값**이라고 합니다.
* pandas 에서는 NaN => Not a Number 로 표기 된 것을 확인해 볼 수 있습니다.

```python
df
```

info() 로 NaN 값, 즉 빠진 데이터가 어디에 있는지 쉽게 요약정보로 확인할 수 있습니다.

```python
df.info()
```

## 5.2. 결측 값 다루기

- df.isna(), df.isnull() : Nan 유무를 boolean으로 반환
- isnull()과 isna() 는 같은 기능

```python
df.isna()
```

- df[column명].isnull() : column별로 Nan 유무를 알수 있다.

```python
df['그룹'].isnull()
```

```python
df[df['그룹'].isnull()]
```

## 5.3. NaN 값만 색출해내기

```python
df[df['그룹'].isnull()]['그룹']
```

```python
df[df['그룹'].isna()]['그룹']
```

```python
df[df['그룹'].isna()][['그룹','이름']]
```

## 5.4. NaN이 아닌 값에 대하여 Boolean 인덱싱

```python
df['그룹'].notnull()
```

- NaN이 아닌 값만 색출해내기

```python
df[df['그룹'].notnull()]['그룹']
```

```python
df.loc[df['그룹'].notnull(),['키','혈액형']]
```

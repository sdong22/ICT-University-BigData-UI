#### [ 문제1 ] 양의 정수 n을 입력 받아서 아래 그림과 같은 직각삼각형을       출력하는 프로그램을 작성하시오.
[ 조건 ]
 1. 양의 정수 n : 키보드 입력으로 값을 받을 것.
 2. 반복문은 while을 사용할 것
 3. ★(\u2605)을 이용해서 찍을 것.

```python
# 답안1
n = int(input('양의 정수 입력 : '))

i=0

while i < n :
    k= 0
    while k < n-i :
        print('  ', end = '')
        k += 1
    
    
    k= 0
    while k < i+1 :
        print('\u2605', end = '')
        k += 1
            
    
    print()
    i += 1
```

```python
# 답안2
n = int(input('양의 정수 입력 : '))

i=1
k= n-1
while i <= n :
    print('  '*k, '\u2605'*i)
    i+=1
    k -= 1
            
```

```python
# 답안3
n = int(input('양의 정수 입력 : '))

i=1
while i < n+1 :
    print('  '*(n-i)+'\u2605'*(i))
    i += 1
```

#### [ 문제4 ] 파일 입출력

[ 조건 ]

```python
song = ''
with open('test/out.txt','w', encoding='utf8') as outfile:
    outfile.write('')
for i in range(1,5) :
    inFname = 'test/korean_national{}.txt'.format(i)
    with open(inFname,'r', encoding='utf8') as infile:
        song = infile.read()
    with open('test/out.txt','a', encoding='utf8') as outfile:
        outfile.write(inFname+'\n')
        outfile.write('-'*40+'\n')
        outfile.write(song+'\n')
#print(song)    
```

#### [ 문제5 ] 커피
- 갓 볶은 원두로 항상 대기 손님이 있는 맛있는 커피전문점이 있습니다. 제조시간이 많이 걸려서 하루 10잔만 판매하는 '오늘의 커피'는 대기하는 동안 자동 주문 시스템을 통해 주문 가능합니다. 아래 조건에 맞는 시스템을 구축하시오.

[ 조건 ]
 1. 커피 주문은 키보드 입력으로 받을 것.
 2. 대기 손님이 주문할 수 있는 총 오늘의 커피는 10잔으로 재료 소진시
    
    1) 재고가 0이면(10잔 모든 주문이 된 경우)
      : ‘오늘의 커피가 매진되었습니다.’ 메시지 출력후 프로그램 종료.
    
    2) 재고수량보다 많은 량이 주문되면 : '재료가 부족합니다' 메시지 출력 후 프로그램 종료.
 3. 주문 수량에 1보다 작거나 숫자가 아닌 입력값이 들어올때는 ValueError로 처리(ValueError 메시지 : '잘못된 값을 입력하였습니다.')
 4. 커피 유니코드 (\u2615)로 화면 출력(출력예시 참조)

```python
# 답안1
coffee = 10
waiting = 1

while True :
    try :      
        
        print('[남은 커피:{}]'.format(coffee))       
        order = int(input('오늘의 커피 몇 잔 주문하시겠습니까?'))
        
        if order > coffee:
            print('재료가 부족합니다')
            break
        elif order < 1 :
            raise ValueError
        else :
            print('[대기번호{}] 오늘의 커피 '.format(waiting), end='')
            for i in range(order) : print('\u2615',end='')
            print(' 주문이 완료되었습니다.')
            waiting +=1
            coffee -= order
            
        if coffee == 0 : print('오늘의 커피가 매진되었습니다.'); break
    except ValueError :
        print('잘못된 값이 입력되었습니다')
```

```python
# 답안2
class SoldOutError(Exception) : 
    pass

coffee = 10
waiting = 1

while True :
    try:               
        print('[남은 커피:{}]'.format(coffee))
        order = int(input('오늘의 커피 몇 잔 주문하시겠습니까?'))
        
        if order<1 : 
            raise ValueError
        elif coffee < order :
            print('재료가 부족합니다')
            break
        else:
            print('[대기번호{}] 오늘의 커피'.format(waiting),'\u2615'*order,'주문이 완료되었습니다.')
            waiting += 1
            coffee -= order
        
        if coffee ==0 : 
            raise SoldOutError

    except ValueError :
        print('잘못된 값이 입력되었습니다.')
    
    except SoldOutError :
        print('오늘의 커피가 매진되었습니다.')
        break
    
```

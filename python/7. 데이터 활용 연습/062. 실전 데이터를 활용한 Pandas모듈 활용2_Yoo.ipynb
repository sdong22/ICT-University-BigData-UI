# 실전 데이터를 활용한 Pandas모듈 활용2 - 시각화

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#디폴트 폰트가 한글 지원x
plt.rc('font', family='Malgun Gothic') 

# - 부호가 깨질때
plt.rcParams['axes.unicode_minus'] = False

# Graph 사이즈 키우는 옵션
plt.rcParams["figure.figsize"] = (9, 6)
```

```python
# 민간아파트 가격동향
df = pd.read_csv('data/house_price2.csv')
df.info()
```

## 1. line plot
- 주가데이터 처럼 연속성이 있는 데이터들은 보기 좋은데 분양가는 그런 종류가 아니라서 라인 그래프로는 보기가 불편

```python
df['분양가격'].plot(kind='line')
```

```python
df_seoul = df.loc[df['지역명']=='서울']
df_seoul
```

```python
df_seoul_year = df_seoul.groupby('연도').mean()['분양가격']  # ==df_seoul.groupby('연도')['분양가격'].mean()
df_seoul_year
```

```python
df_seoul_year.plot(kind='line')
```

## 2. bar plot

```python
df.groupby('지역명')['분양가격'].mean().plot(kind='bar')
```

```python
df.groupby('지역명')['분양가격'].mean().plot(kind='barh')
```

## 3. histogram

- 도수분포표

```python
df['분양가격']
```

```python
df['분양가격'].plot(kind='hist') 
```

## 4. box plot

```python
df_seoul = df.loc[df['지역명']=='서울']
df_seoul['분양가격'].plot(kind='box')
```

```python
df_seoul.describe()
```

```python
# IQR(Inter Quartile Range) = (3Q-1Q)*1.5
IQR = (7725-6477)*1.5
IQR
```

```python
# box plot max
# Upper Whisker 구간 = 3Q+IQR
7725+IQR
```

```python
# box plot min
# Lower Whisker 구간 = 1Q-IQR
6477-IQR
```

## 5. area plot

```python
df.groupby('월')['분양가격'].count().plot(kind='line', color='pink', alpha=0.7)
df.groupby('월')['분양가격'].count().plot(kind='area', color='pink', alpha=0.2)
```

## 6. pie plot

```python
df.groupby('연도')['분양가격'].count().plot(kind='pie')
```
